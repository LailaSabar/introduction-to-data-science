For this project, I was interestested in using Stack Overflow data from 2017 to inswer those questions : 

  - 1. What are the most used programming languages in United States ?
  - 2. What are the most wanted programming languages in United States ?
  - 3. Are American employees satisfied about their jobs ?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('survey_results_public.csv')
df.head()
```

```python
# explore the hole columns to choose the useful ones to answer our questions
for i in df.columns:
    print (i)
```

```python
#select the useful columns
useful_cols=['Professional','Country','EmploymentStatus','FormalEducation','Gender','Salary','ExpectedSalary','YearsProgram','HaveWorkedLanguage','WantWorkLanguage','JobSatisfaction','YearsCodedJob']
useful_df=df[useful_cols]
useful_df.head()
```

```python
# some information of our columns
useful_df.info()
```

###  1.  What are the most used programming languages in United States ?

```python
# we are interested in the worked languages of programmers from 'United States'
lang_worked=pd.DataFrame(useful_df[useful_df['Country']=='United States']['HaveWorkedLanguage'])
```

```python
# number of null values in the column 'HaveWorkedLanguage' associated to the country USA
lang_worked.isnull().sum()
```

 to handle the missing values we are going to drop the rows concerned, those values are Missing Completely at Random (MCAR) since the survey is designated for developers, and a developer should have a 'HaveWorkedLanguage' value,  so it is safe to remove the missing values

```python
# drop columns with null values
lang_worked.dropna(inplace=True)
lang_worked.isnull().sum()
```

```python

def split_column(df, column):
    ''' Split column by ;,
        returns a splited series.
    '''
    
    df_copy = df
    column_series = df_copy[column].apply(lambda x: x.split(';'))
    
    return column_series
```

```python
# split values in column 'HaveWorkedLanguage' 
worked_lang_splitted = split_column(lang_worked, 'HaveWorkedLanguage')
```

```python
# append every row on the list 'list_lang' and make one common list for all values
list_lang=[]
for i in worked_lang_splitted:
    list_lang.extend(i)
```

```python
# removing any spaces in the begining and the end of a string
list_lang=[x.strip() for x in list_lang]
```

```python
# calculating the percentage of every language
df_lang=pd.DataFrame(list_lang)
lang_count=pd.DataFrame(df_lang[0].value_counts())
lang_count['percentage']=lang_count[0]/(lang_count[0].sum())
lang_count
```

```python
#funtion to plot the result using chart plot
def plot_bar(culumn) :
    plt.figure(figsize=(12,7))
    culumn.plot(kind='bar')
    plt.title('Most Wanted Programming Languages in United States', size = 16)
    plt.ylabel("Percentage", fontsize = 14)
    plt.xlabel("Programming Languages", fontsize = 14)
```

to display our results we will utilize a bar chart to show the changes in a trend. Bar chart are better for comparing larger changes or differences in data among groups.

```python
#plotting the result using chart plot
plot_bar(lang_count['percentage'][:16])
```

After this data exploratory we could easily answer our questions,
by looking at the bars chart above we can notice that the most used programming languages are Javascript in the first place then SQL, followed by python and Java.

```python
# drop columns with null values
lang_want_work.dropna(inplace=True)
lang_want_work.isnull().sum()
```

```python
# append every row on the list 'list_lang2' and make one common list for all values
list_lang2=[]
for i in want_lang_splitted:
    list_lang2.extend(i)
```

```python
# calculating the percentage of every language
df_lang2=pd.DataFrame(list_lang2)
lang_count2=pd.DataFrame(df_lang2[0].value_counts())
lang_count2['percentage']=lang_count2[0]/(lang_count2[0].sum())
lang_count2
```

to display our results we will utilize a bar chart to show the changes in a trend. Bar chart are better for comparing larger changes or differences in data among groups.

the answer to our question is that Americans are interested in learning Javascript, Python, SQL and C#.

### 3.  Are American employees satisfied about their jobs ?

```python
# we are interested in satisfaction degree of American employees
satisfaction_degree=pd.DataFrame(useful_df[useful_df['Country']=='United States']['JobSatisfaction'])
```

```python
# number of null values in the column 'JobSatisfaction' associated to the country USA
lang_want_work.isnull().sum()
```

```python
degree_count=pd.DataFrame(satisfaction_degree['JobSatisfaction'].value_counts()) #count different values of column 'satsfaction' 
degree_count['percentage']=degree_count['JobSatisfaction']/(degree_count['JobSatisfaction'].sum()) #calculate the percentage of everey satisfaction degree
degree_count
```

to display the results we will use pie charts,it is generally used to show percentage or proportional data and usually the percentage represented by each category is provided next to the corresponding slice of pie

```python
# plot the results using pie chart plot
plt.figure(figsize=(12,7))
degree_count['percentage'].plot(kind='pie')
plt.title('Employees job satisfaction in United States', size = 16)
plt.ylabel("Percentage", fontsize = 14)
plt.xlabel("degree of satisfaction", fontsize = 14)
```

By looking at the pie chart above, we notice that most of Americans are highly satisfied about their jobs.

```python
#Coding the different values of column 'YearsCodedJob'

years_coded = {'1 to 2 years' : 1, 
                '10 to 11 years' : 10, 
                '11 to 12 years' : 11, 
                '12 to 13 years' : 12,
                '13 to 14 years' : 13, 
                '14 to 15 years' : 14, 
                '15 to 16 years' : 15, 
                '16 to 17 years' : 16,
                '17 to 18 years' : 17, 
                '18 to 19 years' : 18, 
                '19 to 20 years' : 19, 
                '2 to 3 years' : 2,
                '20 or more years' : 20, 
                '3 to 4 years' : 3, 
                '4 to 5 years' : 4, 
                '5 to 6 years' : 5, 
                '6 to 7 years' : 6, 
                '7 to 8 years' : 7, 
                '8 to 9 years' : 8, 
                '9 to 10 years' : 9, 
                'Less than a year' : 0}
```

 to handle the missing values we are going to drop the rows concerned, those values are Missing Completely at Random (MCAR) since the survey is designated for developers and a developer should have a 'YearsCodedJob' value, so it is safe to remove the missing values

```python
#drop the missing values in column 'YearsCodedJob'
useful_df['YearsCodedJob'].dropna(inplace=True)
```

```python
#Coding the different values of column 'YearsCodedJob' by applying the dictionnary years_coded
useful_df['YearsCodedJob']=useful_df['YearsCodedJob'].apply(lambda x: years_coded[x])
useful_df['YearsCodedJob'].dropna(inplace=True)
```

```python
# counting different values of the column YearsCodedJob
useful_df['YearsCodedJob'].value_counts()

```

to display our result we will use a line graph, it is used to track changes over short and long periods of time. When smaller changes exist.

```python
# plot the results 
useful_df['Salary'].groupby(useful_df['YearsCodedJob']).mean().plot()
plt.title("How much experience increase salary in United States");
plt.xlabel('YearsCodedJob')
plt.ylabel('Average Salary')
```

We notice that experience impact salaries, which means employees with long experience are earn the highest salaries
